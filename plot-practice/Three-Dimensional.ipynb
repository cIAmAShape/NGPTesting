```python
#Explicit Surfaces (based on f(x,y) = z format ONLY)
ax3 = plt.axes(projection='3d')
x_data3 = np.arange(-5, 5, 0.1)
y_data3 = np.arange(-5, 5, 0.1)
X3, Y3 = np.meshgrid(x_data3, y_data3)
Z3 = np.sin(X3) + np.cos(Y3) * np.sin(Y3)
ax3.plot_surface(X3, Y3, Z3, cmap='plasma')
```

```python
#Implicit 3d-plotting for a sphere
ax4 = plt.axes(projection='3d')

def f(x,y,z):
    return x**2 + y**2 + z**2 - 1  #sphere of radius 1

x4, y4 = np.linspace(-2, 2, 40), np.linspace(-2, 2, 40)
X4, Y4 = np.meshgrid(x4, y4)
z_values = np.linspace(-2, 2, 10)

index = 1
ax4.set_xlabel('X axis')
ax4.set_ylabel('Y axis')
ax4.set_zlabel('Z axis')
for z in z_values:
    Z = f(X4, Y4, z)
    ax4.contour(X4, Y4, Z, levels=[x for x in range(-10, 30)])
#ax4.view_init(elev=35, azim=45)  # Change elevation and azimuth
#ax4.set_zlim([-5, 30])
```
