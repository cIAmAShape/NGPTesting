```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_arr = np.linspace(0, 15, 500)
```

```python
y0 = np.sin(x_arr)
y1 = np.cos(x_arr)
```

```python
y0_noise = 0.05 * np.random.randn(len(x_arr))
y1_noise = 0.05 * np.random.randn(len(x_arr))

y0_sig = y0 + y0_noise
y1_sig = y1 + y1_noise
```

```python
fig, ax = plt.subplots()
ax.plot(x_arr, y0_sig, label='Sine Wave w/ Noise', color='purple')
ax.plot(x_arr, y1_sig, label='Cosine Wave w/ Noise', color='blue')
```
